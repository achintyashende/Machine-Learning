```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Position_Salaries.csv")
```

```python
data
```

## Splitting data

```python
x = data.iloc[:, 1:-1].values
y = data.iloc[:,-1].values
```

```python
x
```

```python
y
```

## Linear Regression Model

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(x,y)
```

## Visualise LR Model

```python
plt.scatter(x,y)
plt.plot(x, lin_reg.predict(x), color = "red")
plt.show()
```

# <font color ="green">Ploynomial Linear Regression</font>

```python
from sklearn.preprocessing import PolynomialFeatures

ploy_reg = PolynomialFeatures(degree = 2)
x_poly = ploy_reg.fit_transform(x)
```

```python
x_poly
```

```python
poly_linear_reg = LinearRegression()
poly_linear_reg.fit(x_poly,y)
```

## Visualising Ploynomial Linear Regression

```python
plt.scatter(x,y)
plt.plot(x, poly_linear_reg.predict(x_poly), color = "red")
plt.title("Ploynomial LR")
plt.xlabel("Input")
plt.ylabel("Predection")
plt.show()
```

```python
ploy_reg = PolynomialFeatures(degree = 3)
x_poly = ploy_reg.fit_transform(x)
poly_linear_reg = LinearRegression()
poly_linear_reg.fit(x_poly,y)
plt.scatter(x,y)
plt.plot(x, poly_linear_reg.predict(x_poly), color = "red")
plt.title("Ploynomial LR")
plt.xlabel("Input")
plt.ylabel("Predection")
plt.show()
```

```python
ploy_reg = PolynomialFeatures(degree = 4)
x_poly = ploy_reg.fit_transform(x)
poly_linear_reg = LinearRegression()
poly_linear_reg.fit(x_poly,y)
plt.scatter(x,y)
plt.plot(x, poly_linear_reg.predict(x_poly), color = "red")
plt.title("Ploynomial LR")
plt.xlabel("Input")
plt.ylabel("Predection")
plt.show()
```
